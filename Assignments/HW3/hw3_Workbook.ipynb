> __a)__ Spark is high-performance, general-purpose distributed computing system that is an open-source project from Apache. Spark accomplishes these feats without over-burdening the developer with the complexities of parallelization and fault tolerance. Like Hadoop MapReduce, Spark offers a general purpose API that is based off of the functional programming paradigm.  But Spark differs from Hadoop MapReduce in a few important ways. > __c)__ Spark's DAG (directed acyclical graph) is the execution plan for a Spark job, representing all of work to be performed for a given action. By evaluating the DAG, a Spark developer can gain insight into how the tasks (transformations) are parallelized and importantly, where shuffles are occuring.  Decomposing the work down to this detail will help the developer how to make the Spark job more performant by minimizing the shuffles and increasing the parallelization where possible. 

> __d)__ We want to cache an RDD when we want to perform different operations on the same output of on a series of transformations. > __b)__ Like Cosine similarity, the Jaccard, Overlap, and Dice metrics are used to score the similarity between 2 sets.  Cosine, Jaccard and Overlap all have the same numerator, which is the number of attributes that are in both sets. (Dice is multiplies this same numerator times 2.)  However, Cosine similarity can go beyond counting the number of instances in a set, and instead apply weights to each attribute, which can be useful when accounting for term frequency.  
>

>

Dice captures overlap as a percentage, multiplying the denominator by 2 and using the sum of length of set A  and the length of set B as the denominator.  We can think of Jaccard, Overlap and Dice as measuring proportional similarity between documents.
>
> Cosine similarity is a score based on orientation, not magnitude and has the range between -1 and 1, more specifically a range between 0 and 1 for positive dimension.  This is efficient for computations, but doesn't take into consideration, for example, document length which may be important when evaluating 2 sets of very different sizes.  The other metrics (Jaccard, Overlap, and Dice) are helpful when the weight of each attribute is not important and assessing the degree of overlap (quantity rather than quality) is appropriate.

# Question 3: Synonym Detection Strategy

In the Synonym Detection task we want to compare the meaning of words, not documents. For clarity, lets call the words whose meaning we want to compare `terms`. If only we had a 'meaning document' for each `term` then we could easily use the document similarity strategy from Question 2 to figure out which `terms` have similar meaning (i.e. are 'synonyms'). Of course in order for that to work we'd have to reasonably believe that the words in these 'meaning documents' really do reflect the meaning of the `term`. For a good analysis we'd also need these 'meaning documents' to be fairly long -- the one or two sentence dictionary definition of a term isn't going to provide enough signal to distinguish between thousands and thousands of `term` meanings.

This is where the idea of co-occurrance comes in. Just like DocSim makes the assumption that words in a document tell us about the document's meaning, we're going to assume that the set of words that 'co-occur' within a small window around our term can tell us some thing about the meaning of that `term`. Remember that we're going to make this 'co-words' list (a.k.a. 'stripe') by looking at a large body of text. This stripe is our 'meaning document' in that it reflects all the kinds of situations in which our `term` gets used in real language. So another way to phrase our assumption is: we think `terms` that get used to complete lots of the same phrases probably have related meanings. This may seem like an odd assumption but computational linguists have found that it works surprisingly well in practice. Let's look at a toy example to build your intuition for why and how.

__Document Similarity Analysis in Spark:__

````python
# load data - RUN THIS CELL AS IS
data = sc.textFile(hw3_path+"sample_docs.txt")  
````

````python
# helper function - RUN THIS CELL AS IS
def splitWords(pair):
    """Mapper 2: tokenize each document and emit postings."""
    doc, text = pair
    words = text.split(" ")
    for w in words:
        yield (w, [(doc,len(words))])
````

````python
# helper function - RUN THIS CELL AS IS
def makeCompositeKey(inverted_index):
    """Mapper 3: loop over postings and yield pairs."""
    word, postings = inverted_index
    # taking advantage of symmetry, output only (a,b), but not (b,a)
    for subset in itertools.combinations(sorted(postings), 2):
        yield (str(subset), 1)
````

````python
# helper function - RUN THIS CELL AS IS
def jaccard(line):
    """Mapper 4: compute similarity scores"""
    (doc1, n1), (doc2, n2) = ast.literal_eval(line[0])
    total = int(line[1])
    jaccard = total / float(int(n1) + int(n2) - total)
    yield doc1+" - "+doc2, jaccard
````

````python
# Spark Job - RUN THIS CELL AS IS
result = data.map(lambda line: line.split('\t')) \
             .flatMap(splitWords) \
             .reduceByKey(lambda x,y : x+y) \
             .flatMap(makeCompositeKey) \
             .reduceByKey(lambda x,y : x+y) \
             .flatMap(jaccard) \
             .takeOrdered(10, key=lambda x: -x[1])
   
result
````

> __b)__ To compute the possible number of pairs, we use the combinations equation:

> __c)__ The most frequent words are common parts of speech and have little inherent semantic value.  As such, this will produce noise, flooding out possible meaningful patterns to detect similar documents.

> __d)__ The least frequent words have very specific meanings, as such, we are unlikely to encounter these words in new documents we encounter.  So if we don't include a large enough set of infrequent words across many training documents, we run the risk of not detecting any overlap.

````python
# part a - write your spark job here 
def EDA1(rdd, n):
    total, top_n, bottom_n = None, None, None
    ############# YOUR CODE HERE ###############
    
    # helper function
    # split the ngram, which is the first column
    # and emit each word and the word count
    def splitNGram(ngram_rec):
      ngram, word_count = ngram_rec
      word_count = int(word_count)
      words = ngram.split(" ")
      for w in words:
          yield (w, word_count)
    
    # Run the Spark job to 
    # 1. split the ngram records,
    #    emitting each word in the ngram and the word count.
    # 2. sum word counts by key
    word_freq = rdd.map(lambda line: line.lower().split('\t')[:2]) \
       .flatMap(splitNGram) \
       .reduceByKey(lambda x,y : x+y) \
       .cache()
    
    # Grab the top n words
    top_n    = word_freq.takeOrdered(n, key = lambda x: -x[1])
    
    # Grab the bottom n words
    bottom_n = word_freq.takeOrdered(n, key = lambda x: x[1])
    
    # Sum the word counts
    total = word_freq.count()
    
    ############# (END) YOUR CODE ##############
    return total, top_n, bottom_n
````

### Q6 Student Answers:

> __b)__ We observe a skewed distribution, where a few pairs are seen a large number of times, but almost all other words drop off in frequency almost immediately, with a long tail distribution.  So this tells us that the majority words have few co-occurences and we will need a fairly large number of features (a mixture of common and uncommon words) to detect similarity.

> __c)__ 88% of the words with the most neighbors are the most frequent words, whereas 1.9% of the words with the fewest neighbors are in th least frequent (rare) words.  This makes sense in terms of the distribution in that the most frequent words are going to have the highest number of neighbors.  In contrast, because of the long tail distribution, words with the least frequent neighbors is vast and only capture a small percentage of the rare words.

````python
# part a - spark job
def EDA2(rdd,n):
    top_n, bottom_n, sampled_counts = None, None, None
    ############# YOUR CODE HERE ###############


    # helper function
    # split the ngram, which is the first column
    # and emit each word and the word count
    def getNGramCombinations(ngram_rec):
      ngram, word_count = ngram_rec
      words = set(ngram.split(" "))
      for idx, word in enumerate(words):
        for cidx, co_word in enumerate(words):
          if idx != cidx:
            yield ( (word, co_word), 1 )

    # Run the Spark job to 
    # 1. split the ngram records,
    #    emitting each word with each neighbor in the ngram
    # 2. get rid of duplicates (word and co-word appearing more than once)
    # 3. now drop the co-word from the key
    # 4. the count for the word represents the number of unique neighbors
    word_combination_freq = rdd.map(lambda line: line.lower().split('\t')[:2]) \
       .flatMap(getNGramCombinations) \
       .distinct() \
       .map(lambda x: (x[0][0], 1)) \
       .reduceByKey(lambda x,y: x+y) \
       .cache()

    # Grab the top n word combinations
    top_n    = word_combination_freq.takeOrdered(n, key = lambda x: -x[1])
    
    # Grab the bottom n word combinations
    bottom_n = word_combination_freq.takeOrdered(n, key = lambda x: x[1])
    
    # Grab n samples of the word combinations
    num_samples = int(rdd.count())
    sampled   = word_combination_freq.takeSample(False, int(np.ceil(num_samples * .01)))
    sampled_counts =  [item[1] for item in sampled]
           
    
    ############# (END) YOUR CODE ##############
    return top_n, bottom_n, sampled_counts
````

````python
# part c - spark job
def get_vocab(rdd, n_total, n_basis):
    vocab, basis = None, None
    ############# YOUR CODE HERE ###############

    stopwords_broadcast = sc.broadcast(set(STOPWORDS))
    
    # helper function
    # split the ngram, which is the first column
    # and emit each word and the word count
    def splitNGram(ngram_rec):
      ngram, word_count = ngram_rec
      word_count = int(word_count)
      words = ngram.lower().split(" ")
      for word in words:
        yield (word, word_count)

    # Run the Spark job to 
    # 1. split the ngram records,
    #    emitting each word that is not a stop word
    # 2. the count for the words
    word_freq = rdd.map(lambda line: line.lower().split('\t')[:2]) \
       .flatMap(splitNGram) \
       .filter(lambda x: x[0] not in stopwords_broadcast.value) \
       .reduceByKey(lambda x,y: x+y) \
       .cache()    
    
  
    # Grab the top n words to form the vocab
    vocab = word_freq.takeOrdered(n_total, key = lambda x: -x[1])
    
    # Grab the bottom n words to form the basis
    # that we will use as features for co-ocurring words
    #basis = word_freq.takeOrdered(n_basis, key = lambda x: x[1])    
    basis = vocab[-1000:]
    
    
    ############# (END) YOUR CODE ##############
    return vocab, basis
````

````python
# part d - spark job
def buildStripes(rdd, vocab, basis):
    stripesRDD = None
    ############# YOUR CODE HERE ###############
    

    vocab_bset = sc.broadcast(set([x[0] for x in vocab]))
    basis_bset = sc.broadcast(set([x[0] for x in basis]))
    
    # helper function
    def stripWords(ngram_rec):
      ngram, word_count  = ngram_rec
      words = ngram.split(" ")
      for idx, term in enumerate(words):
        neighbors = set()
        for nidx, neighbor in enumerate(words):
          if nidx != idx:
            neighbors.add(neighbor)
        filtered_neighbors = neighbors.intersection(basis_bset.value)
        if len(filtered_neighbors) > 0:
          yield (term, filtered_neighbors)

    
    stripesRDD = rdd.map(lambda line: line.lower().split('\t')[:2]) \
       .flatMap(stripWords) \
       .filter(lambda x: x[0] in vocab_bset.value) \
       .reduceByKey(lambda x,y: x | y ) \
       .cache()

    
    ############# (END) YOUR CODE ##############
    return stripesRDD
````

````python
# part d - write your spark job in the space provided
def similarityAnalysis(stripesRDD, n):
    """
    This function defines a Spark DAG to compute cosine, jaccard, 
    overlap and dice scores for each pair of words in the stripes
    provided. 
    
    Output: an RDD, a list of top n, a list of bottom n
    """
    simScoresRDD, top_n, bottom_n = None, None, None
    
    ############### YOUR CODE HERE ################
   

    # helper function
    def splitStripe(stripe):
      """ tokenize each word and emit postings of neighbors."""
      word, neighbors = stripe
      neighbor_count = len(neighbors)
      if neighbor_count > 0:
        for neighbor_word in neighbors: 
          yield (neighbor_word, [(word,neighbor_count)])

    # helper function 
    def makeCompositeKey(inverted_index):
        """ invert index pattern - loop over postings and yield pairs."""
        word, postings = inverted_index
        # taking advantage of symmetry, output only (a,b), but not (b,a)
        for subset in itertools.combinations(sorted(postings), 2):
          if subset[0][0] != subset[1][0]:
            yield (str(subset), 1)
            
            
    # helper function
    def scoreSimilarity(line):
        """ calculate cosine, jaccard, overlap, and dice similarity for co-ocurring word pair"""
        (term1, n1), (term2, n2) = ast.literal_eval(line[0])
        total = int(line[1])
        cosine_score  = cosine( n1, n2, total)
        jaccard_score = jaccard(n1, n2, total)
        overlap_score = overlap(n1, n2, total)
        dice_score    = dice(n1, n2, total)
        yield (term1+" - "+term2, [cosine_score, jaccard_score, overlap_score, dice_score])

    # helper function 
    def jaccard(n1, n2, total):
        """ calculate jaccard similarity"""
        return total / (float(int(n1) + int(n2) - total))

    #def cosine(n1, n2, total):
    #    return total/float( np.sqrt(n1*n1) * np.sqrt(n2*n2) )

    # helper function 
    def cosine(n1, n2, total):
        """ calculate cosine similarity"""
        return total / float( np.sqrt( int(n1) * int(n2) ) )

    # helper function 
    def overlap(n1, n2, total):
        """ calculate overlap similarity"""
        return total / min(n1, n2)

    # helper function 
    def dice(n1, n2, total):
        """ calculate dice similarity"""
        return  (total * 2) / (float(int(n1) + int(n2)))

    result = stripesRDD.flatMap(splitStripe) \
                 .reduceByKey(lambda x,y : x + y) \
                 .flatMap(makeCompositeKey) \
                 .reduceByKey(lambda x,y : x+y) \
                 .flatMap(scoreSimilarity) \
                 .cache()
   
    # Grab the top n word combinations
    top_n    = result.takeOrdered(n, key = lambda x: -x[1][0])
    
    # Grab the bottom n word combinations
    bottom_n = result.takeOrdered(n, key = lambda x: x[1][0])
    
    ############### (END) YOUR CODE ##############
    return result, top_n, bottom_n
````
